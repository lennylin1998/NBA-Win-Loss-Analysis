```python
%matplotlib inline
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
path = r'C:\Users\user\Downloads\chromedriver_win32\chromedriver.exe'
browser = webdriver.Chrome(executable_path=path)
#LAC all season games box score 
url = 'https://www.nba.com/stats/search/team-game/?sort=GAME_DATE&dir=1'
browser.get(url)
browser.find_element_by_xpath('//*[@id="onetrust-accept-btn-handler"]').click()
browser.find_element_by_xpath('//*[@id="custom-filters"]/div/table/tbody/tr/td[4]/button/i').click()
browser.find_element_by_xpath('//*[@id="streak-finder"]/div[2]/section[1]/div[2]/div[2]/button/span').click()
browser.find_element_by_xpath('//*[@id="streak-finder"]/div[2]/section[1]/div[3]/div/form/div/section[1]/div/div[3]/div/div[4]').click()
browser.find_element_by_xpath('//*[@id="streak-finder"]/div[2]/section[1]/div[3]/div/form/div/section[1]/div/div[2]/ul/li[3]').click()
browser.find_element_by_xpath('//*[@id="streak-finder"]/div[2]/section[1]/div[2]/div[2]/querytool-run-it/button').click()
time.sleep(3)
browser.fullscreen_window()
```

```python
#creating teams_list for iterating
browser.execute_script("window.scrollTo(0, 0)") 
browser.find_element_by_xpath('//*[@id="streak-finder"]/div[2]/section[1]/div[2]/div[2]/button/span').click()
browser.find_element_by_xpath('//*[@id="streak-finder"]/div[2]/section[1]/div[3]/div/form/div/section[2]/div/div[3]/div/div[4]').click()
teams = browser.find_elements_by_xpath('/html/body/main/div/div/div[2]/section[1]/div[3]/div/form/div/section[2]/div/div[2]/ul//li[@class="multiselect-box__option"]')
teams_list=[]
for team in teams:
    #print(team.text)
    teams_list.append(team.text)
```

```python
#parsing data with bs4 for all teams

final_data=[]
i=0
for team in teams_list:
    browser.find_element_by_xpath('//*[@id="streak-finder"]/div[2]/section[1]/div[2]/div[2]/button/span').click()
    browser.find_element_by_xpath('//*[@id="streak-finder"]/div[2]/section[1]/div[3]/div/form/div/section[2]/div/div[3]/div/div[4]').click()
    browser.find_element_by_xpath('//*[@id="streak-finder"]/div[2]/section[1]/div[3]/div/form/div/section[2]/div/div[2]/ul/li[2]').click()
    if i==0:
        i+=1
        pass
    else:
        browser.find_element_by_xpath('//*[@id="streak-finder"]/div[2]/section/div[3]/div/form/div/section[2]/div/div[2]/ul/li[1]/span/i').click()
    browser.find_element_by_xpath('//*[@id="streak-finder"]/div[2]/section[1]/div[3]/div/form/div/section[2]/div/div[2]/ul/li[1]/input').send_keys(team)
    browser.find_element_by_xpath('//*[@id="streak-finder"]/div[2]/section[1]/div[3]/div/form/div/section[2]/div/div[2]/ul/li[3]').click()
    browser.find_element_by_xpath('//*[@id="streak-finder"]/div[2]/section[1]/div[2]/div[2]/querytool-run-it/button').click()
    browser.execute_script("window.scrollTo(0, 500)")
    browser.find_element_by_xpath('//*[@id="streak-finder"]/div[2]/section[2]/querytool-stat-table/div[2]/div/button').click()
    browser.execute_script("window.scrollTo(0, 100)")
    soup = BeautifulSoup(browser.page_source,'html')
    tables = soup.find_all('table')
    table = tables[0]
    data = [[info.text for info in row.find_all(["th",'td'])] for row in table.find_all("tr")]
    print(data)
    final_data.append(data)
    data=[]
    time.sleep(2)
    
print(len(final_data))
```

```python
#inspecting and cleaning data
#print(final_data)
df=pd.DataFrame(final_data[0])
for n in range(1, len(final_data)):
    df = df.append(pd.DataFrame(final_data[n]),ignore_index=True)
df.columns = df.iloc[0 ,:]
df.drop(index=0, inplace=True)
index = df[df["TEAM"]=="TEAM"].index
df.drop(index=index, inplace=True)
df["TEAM"] = df["TEAM"].apply(lambda x: x.replace('\n', ''))
df["HOME"] = df["MATCHUP"].apply(lambda x: True if "vs." in x else False)
df["MATCHUP"] = df["MATCHUP"].apply(lambda x: x[-3:])
print(df.head())
print(df.info())
df.to_excel("NBA_2020-21.xlsx")
```
